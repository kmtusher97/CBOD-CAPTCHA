```python
import os
import math
import random
import numpy as np
from PIL import Image, ImageDraw
```

```python
filename = "test-img657540.jpeg"
img = Image.open('./captcha/' + filename)
display(img)
```

```python
def random_pixel(x_l, x_r, y_l, y_r):
    return (int(random.uniform(x_l, x_r)), int(random.uniform(y_l, y_r)))

random_pixel(-10, 5, 5, 15)
```

```python
def generate_random_polygon(center_p):
    x, y = center_p
    poly = []
    poly.append(random_pixel(x - 10, x + 10, y - 45, y - 70))
    poly.append(random_pixel(x + 45, x + 70, y - 10, y + 10))
    poly.append(random_pixel(x - 10, x + 10, y + 45, y + 70))
    poly.append(random_pixel(x - 45, x - 70, y - 10, y + 10))
    return poly

generate_random_polygon((50, 50))
```

```python
colors = ["red", "green", "blue", "yellow", "black", "brown", "orange"]
centers = [(50, 50), (150, 50), (50, 150), (150, 150)]
polygons =[]
for center_i in centers:
    polygons.append(generate_random_polygon(center_i))

img1 = ImageDraw.Draw(img)
for poly in polygons:
    n = len(poly)
    clr = colors[int(random.uniform(0, len(colors) - 1))]
    for i in range(n):
        img1.line([poly[i], poly[(i + 1) % n]], fill = clr, width = int(random.uniform(2, 4)))

display(img)
```

```python
img.save('./captcha/' + filename + "-bb." + img.format.lower())
```
